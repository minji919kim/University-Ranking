```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd

#2022-2023
ranking_num = []
uni_name = [] 
country = []
national_rank = []
educ_rank = []
employ_rank = [] 
faculty_rank = []
research_rank = [] 
score = [] 

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

url = "https://cwur.org/2022-23.php" #Data from SlickCharts
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.text, 'html.parser')

rows = soup.find_all('tr')
for row in rows:
    all_td_tags = row.find_all('td')
    if len(all_td_tags) > 0:
        ranking_num.append(all_td_tags[0].text)
        uni_name.append(all_td_tags[1].text)
        country.append(all_td_tags[2].text)
        national_rank.append(all_td_tags[3].text)
        educ_rank.append(all_td_tags[4].text)
        employ_rank.append(all_td_tags[5].text)
        faculty_rank.append(all_td_tags[6].text)
        research_rank.append(all_td_tags[7].text)
        score.append(all_td_tags[8].text)
        
# initialize data of lists.
data = {'World Rank': ranking_num,
        'Institution': uni_name, 
       'Location': country, 
       'National Rank': national_rank, 
       'Education Rank': educ_rank, 
       'Employability Rank': employ_rank, 
       'Faculty Rank': faculty_rank, 
       'Research Rank': research_rank, 
       'Score': score, 
       'Year': 2023}
  
# Creates pandas DataFrame.
df22_23 = pd.DataFrame(data)
df22_23.to_csv('/Users/minjikim/Desktop/Data Portofolio/University Ranking/University_Ranking_2023.csv', sep=',') 
```
